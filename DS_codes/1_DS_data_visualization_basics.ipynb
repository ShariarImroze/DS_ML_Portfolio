A basic implementaion of seaborn library. 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

Analysis of FIFA dataset. 

```python
fifa_filepath = 'DS_assets/1_soccer_data.csv'

fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates =True)
```

```python
fifa_data.head()
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(data=fifa_data)
```

Analysis of daily streams on the the music streaming service Spotify. 

```python
spotify_filepath = 'DS_assets/1_spotify.csv'

spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates = True)
```

```python
spotify_data.head()
```

```python
spotify_data = spotify_data.dropna()
```

```python
spotify_data.head()
```

```python
plt.figure(figsize=(20,10))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data)
```

```python
plt.figure(figsize=(20,10))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")
sns.lineplot(data=spotify_data['Despacito'],label="Despacito")
plt.xlabel("Date")
```

Analysis of Flight Data

```python
flight_filepath = 'DS_assets/1_delays_flight.csv'

#check by renaming to flight_delays

flight_data = pd.read_csv(flight_filepath, index_col="Month") 
```

```python
flight_data.head()
```

```python
plt.figure(figsize=(20,10))

plt.title("Average Arrival Delay for Sprit Airlines Flights (Monthly)")

sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(20,10))

plt.title("Average Arrival Delay for Each Airline (Monthly)")

sns.heatmap(data=flight_data, annot=True)

plt.xlabel("Airline")
```

Analysis of Cancer dataset. 

```python
insurance_filepath = 'DS_assets/1_insurance.csv'

insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
insurance_data.columns
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
plt.figure(figsize=(20,10))
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
plt.figure(figsize=(6,6))
color_palette = {'yes': 'red', 'no': 'green'}
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'],palette=color_palette)
```

Analysis of Irish dataset. 

```python
iris_filepath = 'DS_assets/1_iris.csv'
iris_data = pd.read_csv(iris_filepath, index_col="Id")
```

```python
iris_data.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
plt.figure(figsize=(9,5))
sns.set_style("dark")
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
plt.figure(figsize=(9,5))
plt.title("Histogram of Petal Lengths, by Species")
sns.set_style("ticks")
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
```

```python
plt.figure(figsize=(9,5))
plt.title("Distribution of Petal Lengths, by Species")
sns.set_style("ticks")
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
```

Seaborn has five different themes: (1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
